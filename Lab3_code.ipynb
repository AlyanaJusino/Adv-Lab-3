```python
#packages

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import os
```

```python
#Using data uploaded from github to data
github_url = 'https://raw.githubusercontent.com/AlyanaJusino/Adv-Lab-3/refs/heads/main/Phys_Lab_3.csv'

# Using dataframe to create the rows and columns of the data, organizing it by title
df = pd.read_csv(github_url, header=None, names=['T_ms', 'Potval', 'Thermval'])
```

```python
#test print to see if data is correctly written
print(df)
```

```python
#separating into an array for the specific column
Potval = df['Potval'].values
Thermval = df['Thermval'].values
T_ms = df['T_ms'].values

#known values
V_in = 5
R_1 = 1000
R_0 = 3300
T_0 = 298.15
B = 3977
```

```python
#equations
t_s = T_ms /1000

Volt_meas = Thermval * ( V_in / 1023 )

R_t = R_1 * ( V_in / ((Volt_meas) - 1) )

T_k = 1 / ( (1/T_0) + (1/B) * (np.log(R_t/R_0)) )
```

```python
#plotting the title and the potval and thermval

fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(t_s, Potval, color='blue', label='V_0')
ax1.set_xlabel('Time')
ax1.set_ylabel('V_0', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(t_s, Thermval, color='red', linestyle='--', label='A_0')
ax2.set_ylabel('A_0', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Potentiometer and Thermistor Values vs. Time')
```
